# Optimal Experiment Design of Mental Rotation Tasks

## Problem Domain

People are presented with two images, $I_0$ and $I_1$, which are 2D renderings of patterns of $N$ dots. The true configurations are given as $2\times N$ matrices $X_0$ and $X_1$, respectively.  

Participants must determine whether $X_0$ and $X_1$ were generated from the same (albeit possibly transformed and permuted) original pattern of dots, i.e., whether $\exists R,Z\textrm{ s.t. } X_1=ZRX_0$, where $Z$ is a permutation matrix and $R$ is a rotation matrix.

## Computational Model

We can formulate the judgment of whether $I_0$ and $I_1$ have the same origins by deciding about two hypotheses:

* $h_0$: $\forall Z,R\ X_1\neq ZRX_0$
* $h_1$: $\exists Z,R\textrm{ s.t. } X_1=ZRX_0$

To compare the hypotheses, we need to compute the likelihood of each hypothesis, i.e. $\Pr(I_0,\ I_1\ |\ h_0)$ and $\Pr(I_0,\ I_1\ |\ h_1)$.

### Likelihood when the configurations are distinct ($h_0$)

The likelihood is easy to compute under $h_0$, because we assume that $X_0$ and $X_1$ are independent. Thus:

$$\Pr(I_0,\ I_1\ |\ h_0)=\left(\int_{X_0}\Pr(I_0|X_0)\Pr(X_0)\ \mathrm{d}X_0\right)\left(\int_{X_1}\Pr(I_1|X_1)\Pr(X_1)\ \mathrm{d}X_1\right)$$

### Likelihood when the configurations are equivalent ($h_1$)

When the configurations are the same, the likelihood becomes more complicated:

$$
\begin{align*}
\Pr(I_0,\ I_1\ |\ h_1)=&\int_{X_0}\int_{X_1}\int_R\int_Z\Pr(I_0|X_0)\Pr(X_0)\Pr(I_1|X_1)\cdot{}\\\\
&\ \ \ \ \ \ \ \Pr(X_1|X_0,R,Z)\Pr(R)\Pr(Z)\ \mathrm{d}Z\ \mathrm{d}R\ \mathrm{d}X_1\ \mathrm{d}X_0
\end{align*}
$$

For a small number of dots ($N$), we can compute the integral over $Z$ by enumerating every possible mapping between $X_0$ and $X_1$:

$$\Pr(X_1\ |\ X_0,R)=\sum_{Z\in \mathbb{Z}}\Pr(X_1\ |\ X_0,R,Z)\Pr(Z)$$

We can make the assumption that each $Z$ is equally likely, i.e. $\Pr(Z)=\frac{1}{|\mathbb{Z}|}\ $.

$$
\begin{align*}
\Pr(I_0,\ I_1\ |\ h_1)=&\int_{X_0}\int_{X_1}\int_R\Pr(I_0|X_0)\Pr(X_0)\Pr(I_1|X_1)\cdot{}\\\\
&\ \ \ \ \ \ \ \Pr(X_1|X_0,R)\Pr(R)\ \mathrm{d}R\ \mathrm{d}X_1\ \mathrm{d}X_0
\end{align*}
$$

## Implementation
### Some assumptions

Generally, we can describe the process of mental rotation as follows:

1. Observe $I_0$ and $I_1$.
2. (Subconciously) Produce a mental image $M$ from $I_0$.
3. Observe $I_M$.
5. Accept/reject $h_0$ or $h_1$, or go to step 2, replacing $I_0$ with $I_M$.

### Perception

We assume some perceptual uncertainty in the perception $X\rightarrow I$, i.e.

$$\Pr(I\ |\ X)=\pi(I\ |\ X)\approx\mathcal{N}(I\ |\ X, \Sigma)$$

Where the parameter $\Sigma$ is a diagonal matrix representing the variance of the uncertainty in each dimension. We approximate the inverse density, i.e. inferring the true state $X$ from the observation $I$, in a similar manner:

$$\Pr(X\ |\ I)=\pi^{-1}(X\ |\ I)\approx\mathcal{N}(X\ |\ I, \Sigma)$$

### Performing mental rotation (Approach 1)

We cannot literally compute $\int_R\Pr(X_1|X_0,R)\Pr(R)\ \mathrm{d}R\ \ $ from above. Instead, we must use a sequence of mental rotations to approximate $\Pr(X_1|X_0)\ $:

$$
\begin{align*}
\Pr(X_1\ |\ X_0;\alpha_{i},\ldots{},\alpha_0)&=\int_{\mathbf{I_M}}\Pr(X_1|I_{M_{i+1}})\Pr(I_{M_{i+1}}|I_{M_i},\alpha_i)\cdots{}\Pr(I_{M_1}|I_{M_0},\alpha_0)\Pr(I_{M_0}|X_0)\ \mathrm{d}\mathbf{I_M}\\\\
&=\int_{\mathbf{I_M}}\pi^{-1}(X_1|I_{M_{i+1}})\Pr(I_{M_{i+1}}|I_{M_i},\alpha_i)\cdots{}\Pr(I_{M_1}|I_{M_0},\alpha_0)\pi(I_{M_0}|X_0)\ \mathrm{d}\mathbf{I_M}
\end{align*}
$$

Where $\mathbf{I_M}=(I_{M_0}, \ldots{}, I_{M_{i+1}})\ $. To produce a single mental image:

$$
\begin{align*}
\Pr(I_M\ |\ I_0, \alpha)&=\int_{M}\int_{R^{(\alpha)}}\int_X\Pr(I_M\ |\ M)\Pr(M\ |\ R^{(\alpha)},\ X)\Pr(R^{(\alpha)})\Pr(X\ |\ I_0)\Pr(I_0)\ \mathrm{d}M\ \mathrm{d}R^{(\alpha)}\ \mathrm{d}X\\\\
&=\int_{X}\pi(I_M\ |\ M=R^{(\alpha)}X)\ \pi^{-1}(X\ |\ I_0)\ \mathrm{d}X
\end{align*}
$$

#### Choosing the direction of rotation

For each incremental rotation, we must choose a value for $\alpha$. We assume only two options: $\pm\alpha$, where $\alpha$ is actually a small constant (e.g., $\alpha=5^\circ$), and the sign of this angle is given by $d\in[+1,-1]\ \ $.  Using *Expected Information Gain* (EIG), we determine whether $-\alpha$ or $+\alpha$ will give more information about the similarity between $X_0$ and $X_1$:

$$EIG(X_0,\ X_1\ |\ d_i)=H(X_0,\ X_1\ |\ d_{i-1},\ldots{},d_0)-H(X_0,\ X_1\ |\ d_i,d_{i-1},\ldots{},d_0)$$

Where $H$ is *entropy*, defined as:

$$
\begin{align*}
H(X_0, X_1\ |\ d_i,\ldots{},d_0)=&-\int_{X_0}\int_{X_1}\Pr(X_1\ |\ X_0, d_i\alpha,\ldots{},d_0\alpha)\Pr(X_0)\\\\
&\ \ \ \ \ \ \ \ \log\big(\Pr(X_1\ |\ X_0, d_i\alpha,\ldots{},d_0\alpha)\Pr(X_0)\big)\ \mathrm{d}X_0\mathrm{d}X_1
\end{align*}
$$

## Performing mental rotation (Approach 2)

We still want to compute $\int_R \Pr(X_1|X_0, R)\Pr(R)\ \mathrm{d}R$, but we will instead take a Bayesian Numerical Analysis approach. Let $\Pr(R)$ be a uniform Von Mises distribution, such that the integral becomes $\frac{1}{Z}\int_R \Pr(X_1|X_0, R)\ \mathrm{d}R$. Then let $f$ be the probability density function:

$$f(R)=\Pr(X_1|X_0,R)=\int_{I_M}\int_{I_0}\Pr(X_1|I_M)\Pr(I_M|I_0, R)\Pr(I_0|X_0)\ \mathrm{d}I_m\ \mathrm{d}I_0$$

Where:

$$\Pr(I_M|I_0, R)=\int_{X}\pi(I_M\ |\ M=RX)\ \pi^{-1}(X\ |\ I_0)\ \mathrm{d}X$$

The goal then becomes a problem of inferring the value of the integral:

$$g=\int_R f(R)\ \mathrm{d}R=\int_R \Pr(X_1|X_0,R)\ \mathrm{d}R$$

In other words, we want to infer the expected value of $g$ over the posterior distribution of $f$:

$$\hat{g}=\mathbb{E}_f\left[\int_R f\ \mathrm{d}R\ \bigg|\ \alpha_0,\ldots{},\alpha_i,f(\alpha_0),\ldots{},f(\alpha_i)\right]$$

Where:

$$\Pr(f|\alpha_0,\ldots{},\alpha_i)\propto\Pr(f(\alpha_0),\ldots{},f(\alpha_i)|f,\alpha_0,\ldots{},\alpha_i)\Pr(f)$$

The catch, however, is that we cannot choose arbitrary or even necessarily ideal $\alpha$: at each time $t$, we can only choose $\alpha_t\in\\{0,\ \alpha_{t-1}-1,\ \alpha_{t-1}+1\\}$
