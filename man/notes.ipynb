# Optimal Experiment Design of Mental Rotation Tasks

## Problem Domain

People are presented with two images, $I_a$ and $I_b$, which are 2D
renderings of patterns of $N$ dots. The true configurations are given
as $2\times N$ matrices $X_a$ and $X_b$, respectively.

Participants must determine whether $X_a$ and $X_b$ were generated
from the same (albeit possibly transformed and permuted) original
pattern of dots, i.e., whether $\exists R,Z\textrm{ s.t. } X_b=ZRX_a$,
where $Z$ is a permutation matrix and $R$ is a rotation matrix.

## Computational Model

We can formulate the judgment of whether $I_a$ and $I_b$ have the same
origins by deciding about two hypotheses:

* $h_0$: $\forall Z,R\ X_b\neq ZRX_a$
* $h_1$: $\exists Z,R\textrm{ s.t. } X_b=ZRX_a$

The causal structure implied by these hypotheses can be expressed

To compare the hypotheses, we need to compute the likelihood of each
hypothesis, i.e. $ p(I_a,\ I_b\ |\ h_0)$ and $ p(I_a,\ I_b\ |\
h_1)$.

### Likelihood when the configurations are distinct

The likelihood is easy to compute under $h_0$, because we assume that
$X_a$ and $X_b$ are independent. Thus:

$$
\begin{aligned}
 p(I_a,\ I_b\ |\ h_0)&=\left(\int_{X_a} p(I_a|X_a) p(X_a)\ \mathrm{d}X_a\right)\left(\int_{X_b} p(I_b|X_b) p(X_b)\ \mathrm{d}X_b\right)\\
&= p(I_a) p(I_b)
\end{aligned}
$$

### Likelihood when the configurations are equivalent

When the configurations are the same, the likelihood becomes more complicated:

$$ p(I_a,\ I_b\ |\ h_1)=\int_{X_a}\int_{X_b}\int_R\int_Z p(I_a|X_a) p(X_a) p(I_b|X_b) p(X_b|X_a,R,Z) p(R) p(Z)\ \mathrm{d}Z\ \mathrm{d}R\ \mathrm{d}X_b\ \mathrm{d}X_a$$

For a small number of dots ($N$), we can compute the integral over $Z$ by enumerating every possible mapping between $X_a$ and $X_b$:

$$ p(X_b\ |\ X_a,R)=\sum_{Z\in \mathbb{Z}} p(X_b\ |\ X_a,R,Z) p(Z)$$

Making the assumption that each $Z$ is equally likely, i.e. $ p(Z)=\frac{1}{|\mathbb{Z}|}$, we obtain:

$$ p(I_a,\ I_b\ |\ h_1)=\int_{X_a}\int_{X_b}\int_R p(I_a|X_a) p(X_a) p(I_b|X_b) p(X_b|X_a,R) p(R)\ \mathrm{d}R\ \mathrm{d}X_b\ \mathrm{d}X_a$$

This simplifies to:
    
$$ p(I_a,\ I_b\ |\ h1)=\int_R p(I_b|I_a, R) p(I_a) p(R)\ \mathrm{d}R$$

We are now ready to introduce a new variable: $I_R$, denoting a mental image, which is computed deterministically (for now) from $I_a$ and $R$.
    
$$
\begin{aligned}
p(I_a, I_b|h_1)&=\int_R \int_{I_R} p(I_b|I_R) p(I_R|I_a, R)p(I_a)p(R)\ \mathrm{d}I_R\ \mathrm{d}R\\
&= \int_R \int_{I_R} p(I_b|I_R)\delta(RI_a-I_R)p(I_a)p(R)\ \mathrm{d}I_R\ \mathrm{d}R\\
&= \int_R p(I_b|I_R)p(I_a)p(R)\ \mathrm{d}R
\end{aligned}
$$

Assuming all angles are equally likely, this simplifies further to:
    
$$p(I_a, I_b|h_1) \propto \int_R p(I_b|I_R)\ \mathrm{d}R$$

## Algorithmic Implementation

There are two assumptions we make:

1. Computing any single $I_R$ is a costly operation in that it is both a lengthy operation and memory-intensive. Moreover, $I_R$ must be generated sequentially, that is, for some small angle $r$, we must generate $I_{R\pm r}$ before we can generate $I_R$.
2. We cannot know the quantity $p(I_b|I_R)$ exactly, so we approximate it with a pixel-space similarity function $S(I_b, I_R)$. We do not know the analytic form of this function, but we can evaluate it for any $I_b$ and $I_R$.

Thus, the desired quantity for $h_1$ becomes:
    
$$p(I_a,I_b|h_1)\approx \frac{1}{C}\int_R S(I_b, I_R)\ \mathrm{d}R$$

Where $C$ is a normalizing constant. We cannot rely on a simple Monte Carlo simulation to estimate this integral: because mental rotations are costly in terms of cognitive resources and must be performed in a sequence, we cannot evaluate $S(I_b, I_R)$ at arbitrary $I_R$. Instead, we must choose a relatively small, sequential set of $I_R$ to estimate the integral.

### Bayesian Quadrature

We can use what is known as *Bayesian quadrature* \cite{Diaconis:1988uo} to estimate the above integral by first computing a posterior distribution over $S$. We choose $p(S)$ to be a periodic Gaussian process \cite{Rasmussen:2006vz} and let a sequence of $N$ consecutive mental rotations be $\mathbf{R}=\{R_1, \ldots{}, R_N\}$. Then:
    
$$p\big(S\ \big|\ I_b, \mathbf{R}\big)\propto p\big(S\big)\prod_{R\in\mathbf{R}} p\big( S(I_b, I_R)\ \big|\ S, I_b, I_R\big)$$

We can take the MAP of $S$ to obtain a single function:
    
$$\hat{S}=\mathrm{argmax}_S p(S\ |\ I_b, \mathbf{R})$$

This now gives us a form that we can integrate. Thus:

$$p(I_a, I_b | h_1)\propto \int_R p(I_b | I_R)\ \mathrm{d}R\approx \mathrm{MAP}\left[\frac{1}{C}\int_R S(I_b, I_R)\ \mathrm{d}R\ \bigg|\ I_b, \mathbf{R}\right]=\int_0^{2\pi} \hat{S}(I_b, I_R)\ \mathrm{d}R$$

### Choosing the sequence of mental images

The sequence of mental rotations $\mathbf{R}$ given above is constrained by the fact that each rotation $R_i$ is dependent on the previous one:
    
$$R_i\in \{R_{i-1}-r,\ R_{i-1},\ R_{i-1}+r\}$$

Where, as before, $r$ is a small angle. Given this constraint, how do we choose $\mathbf{R}$ such that it will give us the best approximation of $p(I_a, I_b|h_1)$? One option is to choose each $R_i$ in a greedy manner to minimize the variance of $\hat{S}$ (i.e. give a more accurate estimate of $S$).
