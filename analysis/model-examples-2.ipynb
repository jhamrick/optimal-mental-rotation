# Models of mental rotation

```python
%matplotlib inline
```

```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
from path import path

import mental_rotation.model as models
from mental_rotation.stimulus import Stimulus2D
import analysis_util as util

np.seterr(invalid='ignore')
```

```python
ROOT = path("../")
CONFIG = util.load_config(ROOT.joinpath("config.ini"))
VERSION = CONFIG.get("global", "version")
SEED = CONFIG.getint("global", "seed")
STIM_PATH = ROOT.joinpath(CONFIG.get("paths", "stimuli"))
FIG_PATH = ROOT.joinpath(CONFIG.get("paths", "figures"))

opts = {
    'S_sigma': CONFIG.getfloat("model", "S_sigma"),
    'step': CONFIG.getfloat("model", "step")
}
```

```python
def format_plot(fig, ax1, ax2):
    ax1.set_xlim(-np.pi, np.pi)
    ax1.set_xticks([-np.pi, -np.pi/2., 0, np.pi/2., np.pi])
    ax1.set_xticklabels([-180, -90, 0, 90, 180])
    ax1.set_ylabel("Similarity")
    ax1.set_title("Same")
    ax2.set_title("Different")
    for ax in [ax1, ax2]:
        util.clear_right(ax)
        util.clear_top(ax)
        ax.set_xlabel("Rotation")
    fig.set_figwidth(8)
    plt.tight_layout()
```

## Example stimulus

```python
X = Stimulus2D.load(STIM_PATH.joinpath(VERSION, "011_120_0.json"))
Xa = X.copy_from_initial()
Xb = X.copy_from_vertices()

fig, (ax1, ax2) = plt.subplots(1, 2)
Xa.plot(ax1)
Xb.plot(ax2)
```

## Gold standard model

```python
np.random.seed(SEED)
gs_model = models.GoldStandardModel(Xa.vertices, Xb.vertices, **opts)
gs_model.sample()
gs_model.print_stats()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
lines0 = gs_model.plot(ax1, 0)
lines1 = gs_model.plot(ax2, 1)
format_plot(fig, ax1, ax2)
```

## Oracle model

```python
np.random.seed(SEED)
o_model = models.OracleModel(Xa.vertices, Xb.vertices, **opts)
o_model.sample()
o_model.print_stats()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
lines0 = o_model.plot(ax1, 0, f_S=gs_model.S)
lines1 = o_model.plot(ax2, 1, f_S=gs_model.S)
format_plot(fig, ax1, ax2)
```

## Hill-climbing model

```python
np.random.seed(SEED)
hc_model = models.HillClimbingModel(Xa.vertices, Xb.vertices, **opts)
hc_model.sample()
hc_model.print_stats()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
lines0 = hc_model.plot(ax1, 0, f_S=gs_model.S)
lines1 = hc_model.plot(ax2, 1, f_S=gs_model.S)
format_plot(fig, ax1, ax2)
```

## Threshold model

```python
np.random.seed(SEED)
t_model = models.ThresholdModel(Xa.vertices, Xb.vertices, **opts)
t_model.sample()
t_model.print_stats()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
lines0 = t_model.plot(ax1, 0, f_S=gs_model.S)
lines1 = t_model.plot(ax2, 1, f_S=gs_model.S)
format_plot(fig, ax1, ax2)
```

## Bayesian quadrature model

```python
np.random.seed(SEED)
bq_model = models.BayesianQuadratureModel(Xa.vertices, Xb.vertices, **opts)
bq_model.sample()
bq_model.print_stats()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
lines0 = bq_model.plot(ax1, 0, f_S=gs_model.S)
lines1 = bq_model.plot(ax2, 1, f_S=gs_model.S)
format_plot(fig, ax1, ax2)
```

## All models

```python
all_models = [
    ("Oracle model", o_model),
    ("Threshold model", t_model),
    ("Hill-climbing model", hc_model),
    ("Bayesian quadrature model", bq_model)
]
```

```python
xmax = 0
fig, axes = plt.subplots(1, 4, sharey=True, sharex=True)
for (model_name, m), ax in zip(all_models, axes):
    m.plot_trace(ax, legend=(ax == axes[0]))
    
    nstep = m.R_i.size
    if nstep > xmax:
        xmax = nstep

    ax.set_title(model_name, fontsize=14)    
    util.clear_top(ax)
    util.clear_right(ax)
    util.outward_ticks(ax)        
    
axes[0].set_xlim(0, xmax + 1)
axes[0].set_ylabel("Rotation", fontsize=14)

fig.set_figwidth(14)
fig.set_figheight(2)
plt.tight_layout()

util.save(FIG_PATH.joinpath("model-traces-1"), ext=["png", "pdf"], close=False)
```
