```python
def init(n, Ia, Ib, Im, R, opt):
    bq = model_bq.BayesianQuadratureModel(
        Ia[0], Ib[0], Im[:, 0], R,
        **opt)
    bq.fit()
    bq.integrate()

    fig, axes = plt.subplots(1, 3)
    (ax1, ax2, ax3) = axes

    ax2.plot([], [], 'k-', lw=2)
    ax2.plot([], [], 'b-', lw=2)
    ax2.plot([], [], 'b--', lw=2)
    ax2.plot([], [], '--', lw=2, color="#00ff00")
    ax2.set_xlim(-np.pi/12., 2*np.pi + (np.pi / 12.))

    ax3.plot([0], [bq._bq.mean], 'k-', lw=2)
    ax3.plot([0], [bq._bq.mean - np.sqrt(bq._bq.var)], 'k-', alpha=0.5)
    ax3.plot([0], [bq._bq.mean + np.sqrt(bq._bq.var)], 'k-', alpha=0.5)
    ax3.set_xlim(0, n-1)
    fig.set_figwidth(10)
    return fig, axes, bq
```

```python
def plot1(ax, bq, rcurr, rwmin, samp):
    gpmean = bq.S_mean
    gpstd = np.sqrt(bq.S_var)
    lower = gpmean - gpstd
    upper = gpmean + gpstd

    ax.cla()
    ax.set_xlim(-np.pi/12., 2*np.pi + (np.pi / 12.))
    ylim = (-0.1, 1.5)
    ax.set_ylim(*ylim)

    ax.plot([rcurr, rcurr], ylim, 'b-', lw=2)
    if rwmin is not None:
        ax.plot([rwmin, rwmin], ylim, 'b--', lw=2)
    if samp is not None:
        ax.plot([samp, samp], ylim, '--', lw=2, color="#00ff00")
    ax.plot(bq.R, bq.S, 'k-', lw=2)

    ax.fill_between(bq.R, lower, upper, color='r', alpha=0.3)
    ax.plot(bq.R, gpmean, 'r-', lw=2)
    ax.plot(bq._bq.Ri, bq._bq.Si, 'ro', ms=7)
```

```python
def plot2(ax, R, xpc_unc, rcurr, rmin, samp):
    notinf = np.array(xpc_unc)[~np.isinf(xpc_unc)]
    space = np.nanmin(np.clip(notinf, 0, np.inf))
    ylim = np.nanmin(notinf), np.nanmax(notinf) + space    
    ax.set_ylim(*ylim)

    edge = np.ptp(R) * 0.1
    xlim = (np.min(R) - edge, np.max(R) + edge)
    ax.set_xlim(*xlim)

    l, rc, r, s = ax.get_lines()
    l.set_data(R, xpc_unc)
    rc.set_data([rcurr, rcurr], ylim)
    if rmin is None:
        r.set_data([], [])
    else:
        r.set_data([rmin, rmin], ylim)
    if samp is None:
        s.set_data([], [])
    else:
        s.set_data([samp, samp], ylim)
```

```python
def plot3(ax, bq):
    l, ll, lu = ax.get_lines()
    x, y = l.get_data()
    yl = ll.get_data()[1]
    yu = lu.get_data()[1]
    x = np.append(x, x[-1] + 1)
    y = np.append(y, bq._bq.mean)
    l.set_data(x, y)
    yl = np.append(yl, bq._bq.mean - np.sqrt(bq._bq.var))
    ll.set_data(x, yl)
    yu = np.append(yu, bq._bq.mean + np.sqrt(bq._bq.var))
    lu.set_data(x, yu)
    ax.set_ylim(yl.min(), yu.max())
```

```python
def plot(fig, axes, bq):
    rcurr, scurr = bq.curr_val
    rwcurr = rcurr % (2*np.pi)
    d = int(np.round(np.degrees(rcurr)))
    bq.debug("Current: S(%d) = %f" % (d, scurr), level=1)

    dr = bq.opt['dr']
    w = float(bq._bq.gp_S.params[1])
    box = np.max([dr, w])
    r1 = np.radians(1)
    R = np.arange(rcurr - box, rcurr + box + r1, r1)[:, None, None]
    # R = bq.R.copy()[:, None, None]
    Rw = R % (2*np.pi)

    xpc_unc = np.array([
	bq._bq.expected_uncertainty_evidence(r)
	if not bq.observed(r) else np.nan for r in Rw])
    # xpc_unc[xpc_unc <= 0] = np.nan
    min_unc = np.nanargmin(xpc_unc)

    if not np.isnan(min_unc):
        rmin = R[min_unc].ravel()
        rwmin = rmin % (2*np.pi)

        lt = circ.difference(rmin, rcurr) < 0
        gt = circ.difference(rmin, rcurr) > 0
        # if lt > gt:
        #     samp = float(rcurr - np.abs(np.random.normal(dr, w)))
        # elif lt < gt:
        #     samp = float(rcurr + np.abs(np.random.normal(dr, w)))
	samp = rmin

        sampw = samp % (2*np.pi)
        sampwd = float(circ.wrapdiff(np.array([samp], copy=True)))

    else:
        rmin = None
        rwmin = None
        samp = None
        sampw = None
        sampwd = None

    clear_output()

    ax1, ax2, ax3 = axes
    plot1(ax1, bq, rwcurr, rwmin, sampw)
    plot2(ax2, R, xpc_unc, rcurr, rmin, sampwd)
    plot3(ax3, bq)

    plt.subplots_adjust(wspace=0.4)

    return xpc_unc, samp
```
